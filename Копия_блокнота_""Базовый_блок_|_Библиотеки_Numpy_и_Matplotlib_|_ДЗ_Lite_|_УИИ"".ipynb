### Задача 2

Требуется разработать программу для преобразования координат из декартовой системы в полярную.

***Справка***:

* В декартовой системе точка описывается с помощью двух координат (x,y)
* В полярной системе точка описывается с помощью двух координат (полярный угол и полярный радиус)

Для перехода из декартовой системы в полярную необходимо воспользоваться следующими формулами:

- Полярный радиус: $r=\sqrt{x^2 + y^2}$
- Полярный угол: $tg(a)=\frac{y}{x}$, следовательно сам угол можем вычислить следующим образом: $a = arctan(\frac{y}{x})$, где **arctan** - функция арктангенс, которая реализована в библиотеке **numpy** в виде метода `numpy.arctan()` (для x > 0). Если x < 0 угол, стоит считать следующим образом: `np.arctan(y/x)+np.pi`

---

В зависимости от введенных данных ваш результат может быть таким:
pic3.png

```python
# Ваше решение
x=4
y=3
dec_coo=(x,y)
print('Координаты декартовы: ',dec_coo)
pol_rad=((x**2+y**2)**0.5)
# print(pol_rad)
pol_angel=np.arctan(y/x)
# print(pol_angel)
polar_coo=(pol_angel, pol_rad)
print('Координаты полярные: ', ( pol_rad, pol_angel))
```

Используя библиотеку **matplotlib**, постройте графики функций $y = \frac{x^3}{100}$ и $y = x^2$ на одном полотне. При решении задания выполните следующие условия:
- область значений переменной $x$ от -100 до +100 (для генерации значений можно воспользоваться функцией `np.linspace()`);
- задайте название графика;
- задайте подписи к осям;
- задайте подписи (лейблы) для каждого графика функций;
- графики функций должны быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус).

---
Ваш результат должен быть таким:

l5.png

```python
# Ваше решение
import numpy as np
# Кубическая зависимость
x = np.linspace(-100, 100, 50)
y1 = [i**3/100 for i in x]
# Параболическая зависимость
y2 = [i**2 for i in x]

plt.title("Зависимости: y1 = x^3/100, y2 = x^2") 
plt.xlabel("x")         
plt.ylabel("y1, y2")    
plt.grid()              
plt.plot(x, y1, x, y2)  # построение графика
```
